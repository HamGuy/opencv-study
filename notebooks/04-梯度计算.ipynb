##### Scharr 算子 

跟其他的算子就是内核不同，捕捉到更丰富的梯度信息，更加

```python
# 先计算 X
sharrx = cv2.Scharr(grayImage, cv2.CV_64F, 1,0)
sharrx = cv2.convertScaleAbs(sharrx)
# 再计算 Y
sharry = cv2.Sobel(grayImage, cv2.CV_64F, 0,1)
sharry = cv2.convertScaleAbs(sharry)
# 最后计算实际梯度
sharrxy = cv2.addWeighted(sharrx,0.5,sharry,0.5,0)
sharrxy = cv2.convertScaleAbs(sharrxy)

plt.imshow(sharrxy)
```
