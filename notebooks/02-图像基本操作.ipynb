#### ROI - Read Only Intrest

顾名思义，只读取感兴趣的区域，例如，截取图像的部分：

````python
# notebook 里自动绘制图像
%matplotlib inline 

import cv2
import matplotlib
import matplotlib.pyplot as plt

# 读取图像
img = cv2.imread('../imgs/09.png')
# 绘制图像
plt.imshow(img)
````

##### 截取部分图像数据：

对图像数据进行切片操作
```img[sliceH,sliceW]```


````python
#截取竖直和水平方向分别为 1000-2000 的像素点
croped = img[1000:2000,1000:2000]
plt.imshow(croped)
````

##### 颜色通道提取

通过 `cv2.split()`函数提取指定颜色通道的数据

返回值是一个包含当前颜色通道的元组，跟颜色的通道顺序保持一致，默认为 BGR

````python
b,g,r = cv2.split(img)
b
````

````python
# 获取单通道的 shape，同一个图像的各通道 shape 值是相同的
g.shape
````

````python
r.shape
````

##### 图像合并

````python
img2 = cv2.merge((r,g,b))
plt.imshow(img2)
````

##### 只保留 R

````python
imgOnlyR = img2.copy()
# 颜色通道 RGB, 将所有的 G、B 通道赋值为 0
imgOnlyR[:,:,1] = 0
imgOnlyR[:,:,2] = 0

plt.imshow(imgOnlyR)
````

##### 使用 plt 绘制多图

````python
# 绘制多图
plt.subplot(1,2,1)
plt.imshow(img2)
plt.subplot(1,2,2)
plt.imshow(imgOnlyR)
````
