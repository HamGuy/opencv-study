函数原型 `cv.imread(filename[, flags])`

[参数类型](https://docs.opencv.org/4.0.0/d4/da8/group__imgcodecs.html#ga61d9b0126a3e57d9277ac48327799c80)，常用参数如下：
1. cv.IMREAD_COLOR - 加载彩色图像，默认值，也可以直接写0
2. cv.IMREAD_GRAYSCALE - 以灰度图模式加载图像
3. cv.IMREAD_UNCHANGED - 加载彩色图像，包含 alpha 通道

需要注意的是，读取的图像默认是 **BGR** 3通道

```python
# notebook 里自动绘制图像
%matplotlib inline 

import cv2
import matplotlib
import matplotlib.pyplot as plt

# 窗口输出图像
def showImageWindow(title,imgforShow):
    cv2.imshow(title,imgforShow)
    k = cv2.waitKey(0) & 0xFF
    # ESC 键退出
    if k == 27:
        cv2.destroyAllWindows()

# 读取图像
img = cv2.imread('../imgs/09.png')
# 绘制图像
plt.imshow(img)
```

```python
# 灰度图
img2 = cv2.imread('../imgs/09.png',0)
img3 = cv2.imread('../imgs/09.png',cv2.IMREAD_GRAYSCALE)
plt.imshow(img3)
```

```python
# 包含 alpha 通道
img4 = cv2.imread('../imgs/09.png',cv2.IMREAD_UNCHANGED)
plt.imshow(img4)
```
